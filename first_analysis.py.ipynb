```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
import apachelog
```

```python
sample_string = '138.68.19.111 - - [09/Sep/2018:07:16:29 -0400] "GET / HTTP/1.1" 302 5 "http://159.203.94.230:80/" "Mozilla/5.0 zgrab/0.x"'
nformat = r'%h %l %u %t \"%r\" %>s %b \"%i\" \"%{User-Agent}i\"'
p = apachelog.parser(nformat)
data = p.parse(sample_string)
data
```

```python
from apachelog import ApacheLogParserError
log_list = []
with open('access.log') as f:
    for line in f.readlines():
        try:
            data = p.parse(line)
        except ApacheLogParserError:
            sys.stderr.write("Unable to parse %s" % line)
        data['%t'] = data['%t'][1:12]+' '+data['%t'][13:21]+' '+data['%t'][22:27]
        log_list.append(data)
```

```python
#Now lets clean up the pandas dataframe a bit by deleting some columns we aren't interested in and renaming the remaining columns.
```

```python
from pandas import Series, DataFrame, Panel
df = DataFrame(log_list)
df[0:2]
del df['%l']; del df['%u']; del df['%{User-Agent}i']
```

```python
df[0:2]
```

```python
df = df.rename(columns={'%>s': 'Status', '%b':'b', '%h': 'Host', '%i': 'Referer',
                        '%r':'Request', '%t': 'Time'})
```

```python
df[0:2]
```

```python
#Massage the data a bit more so that the access time is a datetime, status is an int, and the response size is in MB, with missing data represented as NaN, not -.
```

```python
df.index = pd.to_datetime(df.pop('Time'))
```

```python
df['Status'] = df['Status'].astype('int')
```

```python
def dash2nan(x):
    if x == '-':
        x = np.nan
    else:
        x = float(x)/1048576.
    
    return x
df['b'] = df['b'].apply(dash2nan)

def dash3nan(x):
    if x == '-':
        x = np.nan
    else:
        x = x
    
    return x
df['Referer'] = df['Referer'].apply(dash3nan)
print(df.head())
```

```python
t_span = '60t'
df_200 = df['Status'][df['Status'] == 200].resample(t_span).agg('count')
df_201 = df['Status'][df['Status'] == 201].resample(t_span).agg('count')
df_202 = df['Status'][df['Status'] == 202].resample(t_span).agg('count')
df_206 = df['Status'][df['Status'] == 206].resample(t_span).agg('count')
df_301 = df['Status'][df['Status'] == 301].resample(t_span).agg('count')
df_302 = df['Status'][df['Status'] == 302].resample(t_span).agg('count')
df_304 = df['Status'][df['Status'] == 304].resample(t_span).agg('count')
df_400 = df['Status'][df['Status'] == 400].resample(t_span).agg('count')
df_404 = df['Status'][df['Status'] == 404].resample(t_span).agg('count')
df_499 = df['Status'][df['Status'] == 499].resample(t_span).agg('count')


status_df = DataFrame({'OK': df_200, 
                       'Resource Created': df_201, 
                       'Request Accepted': df_202, 
                       'Partial Content': df_206, 
                       'Moved Permenantely': df_301, 
                       'Found': df_302, 
                       'Not Modified': df_304, 
                       'Bad Request': df_400,
                       'Not Found': df_404,
                       'Client Closed': df_499})
```

```python
status_df.fillna(0, inplace=True)
status_df[0:5]
```

```python
status_df.plot()
legend()
```

```python

# This doesn't look too bad either, but it's hard to make sense of what is going on with so many different response codes. Let's group them into success (web server handled it as expected) and failure (web server error or client closed request).

# 200    7006    Success    OK 
# 301     564    Success    Moved permanently
# 302     316    Success    Found
# 202     230    Success    Resource Created 
# 201     191    Success    Resource Accepted
# 404      15    Failure    Not Found
# 304      12    Success    Not modified
# 400      11    Failure   Bad request
# 499      10    Failure    Client closed request
# 206       2    Success    Partial content 
```

```python
success_df = df['Status'][df['Status'].isin([200, 301, 302, 201, 202, 304, 206])].resample(t_span).agg('count')

fail_df = df['Status'][df['Status'].isin([404, 400, 499])].resample(t_span).agg('count')
success_df.plot(label="Success")
fail_df.plot(label="Failure")
legend()
```
